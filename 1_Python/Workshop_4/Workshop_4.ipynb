Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

#Имя владельца по номеру документа
def p_people():
    doc_number = input("Введите номер документа: ")          #запрашиваем ввод пользователя
    i = 0                                                    #создаем вспомогательную переменную
    for el in documents:                                     #создаем цикл по списку. Каждый el = словарь   
        i += 1                                               #увеличение значения пермеменной на 1 за каждую итерацию
        if el["number"] == doc_number:                       #проверяем ввод пользователя по ключу "number" словарей el
            print(f'Документ принадлежит: {el["name"]} \n')  #Если условие выполнено, то печатаем вывод по ключевому слову
            break                                            #Прирываем цикл
        elif i == len(documents):                            #Если в ходе цикла перменная i=длине списка - прошли весь список
            print("Документ с запрошенным номером отсутствует") #Пишем, что не нашли запрошенный номер

#Расположение документ по номеру документа
def s_shelf():
    doc_number = input("Введите номер документа: ")
    i = 0
    for el in directories.items():                          #цикл по ключам и значениям словаря. el=список
        i += 1
        if doc_number in el[1]:                             #проверяем есть ли введенный элемент в 1 позиции списка
            print(f"Документ номер {doc_number} находится на полке: {el[0]}")
            break
        elif i == len(directories):
            print("Документ с запрошенным номером отсутствует")

#Вывод списка со всеми документами
def l_list():
    for el in documents:                                       #цикл по всем элементам исходного списка
        print(f'{el["type"]} "{el["number"]}" "{el["name"]}"')

#Создание новой полки        
def add_shelf():
    shelf_number = input("Введите номер новой полки: ")
    if shelf_number in directories:                             #проверяем есть ли введенная полка в ключах исходного словаря
        print("Полка с данным номером уже существует")
    else:
        print(f"Создана новая полка под номером {shelf_number}")
        directories[shelf_number] = []                          #если нет, то добавляет новый ключ с пустым значением в словарь
    print(directories)

#Удаление документа по его номеру
def d_delete():
    doc_number = input("Введите номер документа для удаления: ")
    for el in documents:                                      #цикл по элементам списка
        if el['number'] == doc_number:                        #номер документа по ключевому слову
            documents.remove(el)                              #если номер совпадает, то удалем текущий словарь el из исходного списка 
            for key in directories:                           #создаем ещё один цикл по ключам словаря directories
                if doc_number in directories[key]:            #если в ключах есть номер, то удаляем
                    directories[key].remove(doc_number)
                    print(f"Документ номер {doc_number} успешно удален")
            break
        else:
            print('Документ с запрошенным номером отсутствует')

#Перемещение документа:
def m_move():
    doc_number = input("Введите номер документа для перемещения: ")
    shelf_move = input("Введите новое расположение документа: ")
    for shelf, item in directories.items():                          #цикл с двумя перменными по ключам и значениям словаря
        if doc_number in item and shelf_move in directories.keys():  #если документ есть в значениях словаря и номер в ключах
            directories[shelf].remove(doc_number)                    #то удаляем из текущей полки 
            directories[shelf_move].append(doc_number)               #и добавляем в новую полку
            print(f"Новое расположение документа с номером {doc_number}: {shelf_move}")
            return
    print("Некорректные исходные данные")

#Добавить новый документ в существующую полку    
def a_add():
    doc_type = input("Введите тип нового документа: ")
    doc_number = input("Введите номер нового документа: ")
    doc_name = input("Введите имя владельца номер нового документа: ")
    shelf = input("Введите расположение нового документа: ")
    new_dict = dict(type=doc_type, number=doc_number, name=doc_name)  #создаем вспомогательный словарь на основе введнной информации
    documents.append(new_dict)                                        #добавляем всп. словарь к основному   
    for i in directories.items():                                     #циклом проходим по словарю
        if shelf in directories:                                      #если введенная полка есть в словаре, то     
            directories[shelf].append(doc_number)                     #добавлем в словарь новый номер документа 
            print(f"Новый документ добавлен; расположение нового документа: {shelf}")
            break
        else:
            print("Некорректные исходные данные")
            break

def main():
    print("Имя владельца по номеру документа: p \n"
    "Расположение документ по номеру документа: s \n"
    "Вывод списка со всеми документами: l \n"
    "Создание новой полки: as \n"
    "Удаление документа по его номеру: d \n"
    "Перемещение документа: m \n"
    "Добавить новый документ в существующую полку: a \n"
    "Выход: exit \n")
    while True:
        user_input = input("Введите команду из списка: ")
        if user_input == 'exit':
            print('Завершение программы')
            break
        elif user_input == 'p':
            p_people()
        elif user_input == 's':
            s_shelf()
        elif user_input == 'l':
            l_list()
        elif user_input == 'as':
            add_shelf()
        elif user_input == 'd':
            d_delete()
        elif user_input == 'm':
            m_move()
        elif user_input == 'a':
            a_add()
        else:
            print('Программа не поддерживается')
    
main()
```
