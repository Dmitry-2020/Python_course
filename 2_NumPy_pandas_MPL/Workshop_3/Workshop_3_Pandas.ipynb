# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
  - оценка 2 и меньше - низкий рейтинг
  - оценка 4 и меньше - средний рейтинг
  - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class


```python
def rating_classification(inp):                                 #создаем функцию по условиям задачи
    if inp <= 2:
        return 'low'
    if inp <= 4:
        return 'medium'
    else:
        return 'hight'

rating = pd.read_csv('ratings.csv')                             #считываем исходный файл
rating['class'] = rating.rating.apply(rating_classification)    #применяем функцию к столбцу rating исходной таблицы и создаем новый столбец class с результатами работы функции
rating.head(10)
```

# Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.


```python
keywords = pd.read_csv('keywords.csv')       #считываем исходный файл

geo_data = {                                 #параметры поиска по городам с привязкой к региону
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
keywords.head()
```

```python
def region_search(inp):                           #создаем функцию для поиска
    for region in geo_data:                       #цикл по ключевым словам исходного словаря geo_data
        for city in geo_data[region]:             #цикл по городам исходного словаря geo_data[ключ словаря]
            if city in inp:                       # поиск городов в исходных данных 
                return region
    return 'undefined'

keywords['region'] = keywords['keyword'].apply(region_search) #применяем функцию к столбцу keyword исходной таблицы и создаем новый столбец region с результатами работы функции
keywords.region.value_counts()                      #подсчет уникальных значений по созданному столбцу region
```
