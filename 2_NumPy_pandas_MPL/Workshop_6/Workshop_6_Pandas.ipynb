```python
# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
```

```python
#Исходные данные содержат инофрмацию о квартирах (стоимость, площадь и пр.).
df_initial = pd.read_csv('datasets_71155_151888_flats_moscow.csv')  #импортируем исходные данные
df_initial.head(1)
```

```python
#фильтруем исходные данные, создаем таблицу с нужными столбцами
#'price' - стоимость, 'totsp' - общая площадь, 'kitsp' - площадь кухни, 'dist' - расстояние до центра
df = df_initial[['price', 'totsp', 'kitsp', 'dist']]  
```

```python
#Построение гистограммы: Распределение стоимости квартир

ax1 = df['price'].hist(bins = 50)
ax1.set_title('Стоимость квартиры')
ax1.set_xlabel('Стоимость квартир в 1000$')
ax1.set_ylabel('Количество квартир')
```

```python
#Построение гистограммы: Распределение площади квартир

ax2 = df['totsp'].hist(bins = 50)
ax2.set_title('Общая площадь квартиры')
ax2.set_xlabel('Площадь квартир, кв. м')
ax2.set_ylabel('Количество квартир')
```

```python
#Построение гистограммы: Распределение площади кухни

ax3 = df['kitsp'].hist(bins = 50)
ax3.set_title('Площадь кухни')
ax3.set_xlabel('Площадь кухни, кв. м')
ax3.set_ylabel('Количество квартир')
```

```python
#Построение гистограммы: Распределение расстояния до центра

ax3 = df['dist'].hist(bins = 25)
ax3.set_title('Расстояние от центра')
ax3.set_xlabel('Расстояние от центра, км')
ax3.set_ylabel('Количество квартир')
```

```python
#Используем библиотеку Seaborn для оценки влияния параметров на стоимость квартиры
```

```python
#Зависимость стоимости квартиры от общей площади
sns.jointplot(x='totsp', y='price', 
              data=df, kind='reg')
```

```python
#Зависимость стоимости квартиры от площади кухни
sns.jointplot(x='kitsp', y='price', 
              data=df, kind='reg')
```

```python
#Зависимость стоимости квартиры от расстояния до центра
sns.jointplot(x='dist', y='price', 
              data=df, kind='reg')
```
